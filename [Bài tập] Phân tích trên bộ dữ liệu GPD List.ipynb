```python
import pandas as pd

#đọc dữ liệu, file dữ liệu
data_path = 'C:\\Users\\Qing_Yi99\\codegym\\DAP2308R1\\homework\\'
data_name = 'GDPlist.csv'

data=pd.read_csv(data_path + data_name, encoding= 'unicode_escape')

#hiển thị 5 dòng đầu tiên
data.head()
```

```python
data.info
```

```python
data.rename(columns = {'Country':'Nuoc','Continent':'Chauluc','GDP (millions oGDP (trieu $)'GDP (trieu $)'}, inplace=True)

data
```

```python
#Chèn cột "Thanhpho" vào sau cột "Nuoc" và giá trị ban đầu là của cột "Nuoc"
data.insert(1, 'Thanhpho', pd.Series(data.loc[:,'Nuoc'], index=data.index))
data
```

```python
#Trong cột Thanhpho, thay giá trị Vietnam thành Hanoi; Làm tương tự với các nước còn lại.

data['Thanhpho'].replace('Vietnam', 'Hanoi', inplace=True)
data
```

```python
del data['Thanhpho']
```

```python
#Xóa các bản ghi có Chauluc là ‘Asia’

data.drop(data.loc[data.Chauluc == 'Asia'].index, axis=0, inplace=True)
data
```

```python
#Xóa các bản ghi có GDP < 300000

data.drop(data.loc[data['GDP (trieu $)'] < 300000 ].index, axis=0, inplace=True)
data
```
