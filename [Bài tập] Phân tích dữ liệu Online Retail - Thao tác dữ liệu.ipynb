```python
#Tách cột tháng
retail_df['Month'] = pd.to_datetime(retail_df['InvoiceDate'], format='%m/%d/%Y %H:%M').dt.day
retail_df
```

```python
import numpy as np
```

```python
conditions = [
    (retail_df['Month'].isin([1, 2, 3])),
    (retail_df['Month'].isin([4, 5, 6])),
    (retail_df['Month'].isin([7, 8, 9])),
]
choices = ['1','2','3']
retail_df['Previous'] = np.select(conditions, choices, default='4')
retail_df
```

```python
retail_df['Amount'] = retail_df['Quantity'] * retail_df['UnitPrice']
retail_df
```

```python
conditions =conditions = [
    (retail_df['Country'] == 'United Kingdom') & (retail_df['Previous'] == '4'),
    (retail_df['Country'] == 'France'),
    
]
     
choices = ['10', '5']
retail_df['Discount(%)'] = np.select(conditions, choices, default='0')
retail_df
```

```python
retail_df['Total'] = retail_df['Amount'] - (retail_df['Amount'] * retail_df['Discount(%)'].astype(int)/100)
```
