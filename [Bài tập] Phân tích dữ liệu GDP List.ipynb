```python
#Sử dụng thư viện pandas để làm việc với dữ liệu dạng bảng
import pandas as pd

#đọc dữ liệu, file dữ liệu
data_path = 'C:\\Users\\Qing_Yi99\\codegym\\DAP2308R1\\homework\\'
data_name = 'GDPlist.csv'

data=pd.read_csv(data_path + data_name, encoding= 'unicode_escape')

#hiển thị 5 dòng đầu tiên
data.head()
```

```python
data.shape
print('Data có {} hàng, {} cột'. format(data.shape[0], data.shape[1]))
```

```python
#Thông tin dữ liệu: thang đo tương ứng của các thuộc tính được lưu trữ

data.info()
```

```python
#Biểu đồ GDP các quốc gia

import matplotlib.pyplot as plt

# Xác định trục x và trục y
countries = data['Country']
gdp_values = data['GDP (millions of US$)']

# Vẽ biểu đồ cột (bar chart)
plt.figure(figsize=(12, 6))  # Điều chỉnh kích thước biểu đồ (tùy chọn)
plt.bar(countries, gdp_values)
plt.title("GDP của các quốc gia")
plt.xlabel("Quốc gia")
plt.ylabel("GDP (triệu USD)")

# Để tránh việc tên các quốc gia trên trục x chồng chéo, có thể xoay chúng để dễ đọc hơn.
plt.xticks(rotation=90)

plt.tight_layout()  # Đảm bảo văn bản và nhãn không bị tràn ra ngoài biểu đồ

plt.show()
```

```python
#Mỗi châu lục có bao nhiêu quốc gia nằm trong bảng dữ liệu

total_country = data.groupby('Continent')['Country'].count()
total_country
```
