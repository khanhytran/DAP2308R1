```python
import pandas as pd

#đọc dữ liệu, file dữ liệu
data_path = 'C:\\Users\\Qing_Yi99\\codegym\\DAP2308R1\\homework\\'
data_name = 'OnlineRetail.csv'

data=pd.read_csv(data_path + data_name, encoding= 'unicode_escape')

#hiển thị 5 dòng đầu tiên
data.head(20)
```

```python
#Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia. 

data.pivot_table(values='Quantity', index='InvoiceNo', columns = 'Country',aggfunc ='max')
```

```python
#Xây dựng bảng Pivot table, với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.

data.pivot_table(values='Quantity', index='CustomerID', aggfunc = {min, max})
```

```python
#Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.

data.pivot_table(index='StockCode', values=['Quantity', 'UnitPrice'], aggfunc={'Quantity': 'sum', 'UnitPrice': 'mean'})
```

```python
#Đổi datetime
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
```

```python
# Groupby và tính tổng số lượng hàng bán theo ngày
grouped_data = data.groupby(data['InvoiceDate'])['Quantity'].sum().reset_index()
grouped_data
```

```python
#Tạo pivot table tính tổng lượng hàng bán được mỗi ngày
pivot_table = grouped_data.pivot_table(index='InvoiceDate', values='Quantity')
pivot_table
```
