```python
import pandas as pd

#đọc dữ liệu, file dữ liệu
data_path = 'C:\\Users\\Qing_Yi99\\codegym\\DAP2308R1\\homework\\'
data_name = 'house_price_Dống-Da_Hà-Nội_subdata.csv'

house_df=pd.read_csv(data_path + data_name, encoding= 'unicode_escape')

#hiển thị 5 dòng đầu tiên
house_df.head()
```

```python
#Biến đổi dữ liệu trước khi phân tích
house_df['date'] = pd.to_datetime(house_df['postDate'] - 25_569, unit='d')
```

```python
#Điều chỉnh giá cho chuẩn
'Với các giá ở python, 1 triệu có thể viết: 1_000_000 hoặc 1e6'

house_df['adj_price'] = house_df['price'] * 1e6
```
